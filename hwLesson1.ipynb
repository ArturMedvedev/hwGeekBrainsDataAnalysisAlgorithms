### Задание 1.
Реализовать оптимизацию методом градиентного спуска для x^2(np.sin(0,5_x)^2+1). Какие параметры шага и количества итераций оптимальны? При каких значениях шага оптимизиция не выходит из локальных минимумов? Подробные условия смотрите в .ipynb файле, приложенном к материалам.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def func(x):
    y = (x**2) * ((np.sin(0.5 * x)**2) + 1)
    return y

def grad(x, delta=0.01):
    gradient = (func(x + delta) - func(x - delta)) / (2 * delta)
    return gradient

def true_grad(x):
    gradient = x * (((x * np.sin(x)) / 2) - np.cos(x) + 3)
    return gradient
```

Проверка градиента.

```python
x0 = -40
print(f'Градиент по производной: {true_grad(x0)}     Градиент расчетный: {grad(x0)}')
```

```python
x = np.linspace(-50, 50, 500)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(x, func(x))
plt.show()
```

```python
x0 = np.random.choice(x)
x_next = x0
alfa = 0.05
iter = 0
delta_x = 1e-4
while True:
    iter += 1
    x_next -= alfa * grad(x0)
    if np.abs(x0 - x_next) < delta_x:
        print(f'Итерация: {iter}     x = {x_next}   y = {func(x_next)}')
        break
    x0 = x_next   
```

Опытным путем определил, что при скорости alfa = 0.04 и ниже происходит попадание в локальные минимумы. При скорости alfa = 0.05 локальные минимумы обходятся. Кол-во итераций при этом регулируются точностью, которую задаем параметром delta_x.
